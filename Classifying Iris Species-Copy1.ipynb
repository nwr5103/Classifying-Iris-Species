## Building your first model:k nearest neighbors

```python
import sklearn
print("scikit-learn version: {}".format(sklearn.__version__))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Setting k=1
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

## Making predictions

```python
import numpy as np
print("NumPy version: {}".format(np.__version__))
```

```python
#Predicting the class of an iris  with a sepal length=5cm,width=2.9cm, petal length=1cm and petal width=0.2cm
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]]) 
print("X_new.shape: {}".format(X_new.shape))
```

```python
#Based on measurement, prediction is made for setosa.
```

```python
prediction = knn.predict(X_new) 
print("Prediction: {}".format(prediction)) 
print("Predicted target name: {}".format(
iris_dataset['target_names'][prediction]))
```

## Evaluating the model

```python
# testing accuracy of model by computing fraction of flowers it predicted correctly.
```

```python
y_pred = knn.predict(X_test)
print("Test set predictions:\n {}".format(y_pred))
```

```python

print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python

print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```

```python
#Conclusion: Using this model, 97% of irises predictions were correct.
```

## Strength and Weaknesses of KNN

```python
#Strengths:
#-KNNis an easy algorithm to use.
#-Training data accuracy will always equal 100% because it is based on memory. 
#-There is no computational loss because there is no training.
#-There is no underlying optimization algorithm needed because we do not build a model.
#-It is versatile, it can be used for both classification and regression.

#Weaknesses: 
#-Labelled as a “lazy learner” as there is no explicit learning, it is based on memorization.
#-It can be sensitive to outliers as it can take influence from unique data points. 
#-If you use a small number for k you may be learning uniqueness instead of generalizing. 
#-Since it works based on popular vote, an even number for k can result in a 50/50 split, making it difficult to decide on a predictor.
#-It is not a guaranteed predictor of the best neighbor. As evidenced by the example covered in class with the CIFAR data set, class was the focus but the background was picked up due to pixel intensity.
#- Requires lots of memory to store the training data.
#-Depending on the K we choose it can be a target for under-fitting or overfitting.

```
